```python
 def getSQL(tablename,df):
    result = ''
    file =",".join(df.columns)+",CreateTime,UpdateTime,IsDeleted"
    for index,row in df.iterrows():
        s = ''
        s += 'insert into ' + tablename + ' (' + file + ') values ('
        for i in range(df.columns.size): 
            if str(row[i]) == 'null':
                 rowText= str(row[i])
            elif str(row[i]).find("-")>7:
                rowText= "CAST('"+str(row[i])+"'  AS UNIQUEIDENTIFIER)" if str(row[i]).strip() else '\'\''
            elif df.dtypes[i]=="int64":
                 rowText= str(row[i])
            else:
                 rowText= "'"+str(row[i])+"'"
            s += str(rowText) +','
        s = s[:-1] + ',getdate(),getdate(),0)'
        result += s + '\n'
    return result
```

```python
result = getSQL('REL_V3V5MetaMapping',dfx.fillna(value='null'))
print(result)
with open('InsertData.sql', 'w') as f:
    f.writelines(result)
```
